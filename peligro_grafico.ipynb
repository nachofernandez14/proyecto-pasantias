```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("csv/snic-departamentos-anual.csv", sep=';', encoding='utf-8', low_memory=False)

df = df[df['cod_delito'].isin([15, 16, 19, 20])]

provincia = 'Buenos Aires'
df = df.loc[df['provincia_nombre'] == provincia].copy()

df = df.groupby(['anio'], as_index=False)['cantidad_hechos'].sum()


ao_mas_delitos = df.loc[df['cantidad_hechos'].idxmax()]['anio']
maximo_delitos = df.loc[df['cantidad_hechos'].idxmax()]['cantidad_hechos']
print(ao_mas_delitos, maximo_delitos)
anios = df['anio'].unique()
x_data = df['anio']
y_data = df['cantidad_hechos']

plt.style.use('ggplot')
plt.figure(figsize=(14, 8))
plt.plot(x_data, y_data)
plt.title(f'Cantidad de robos y hurtos en {provincia}')
plt.xticks(anios, rotation=45)
plt.xlabel('Aos')
plt.ylabel('Cantidad de robos/hurtos')
plt.tight_layout()
plt.show()
```
