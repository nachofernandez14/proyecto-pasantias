```python
import pandas as pd
import geopandas as gpd
import unicodedata

df_bsas = pd.read_csv('csv/RADIOS SHEET FINAL_data.csv', encoding='utf-8', sep=';')
gdf = gpd.read_file('shp_radio_censal/radios2022_v1.0.shp')

df_bsas = df_bsas.rename(columns={
    'CÃ³digo de radio': 'radio_codigo',
    'Departamento': 'departamento_nombre',
    'Nivel de incidencia de pobreza crÃ³nica': 'pobreza_cronica_nivel',
    'Provincia': 'provincia_nombre'
})
def normalizar_textos(s):
    if pd.isna(s):
        return ""
    s = s.lower()
    s = "".join(
        c for c in unicodedata.normalize("NFD", s)
        if unicodedata.category(c) != "Mn"
    )
    # Quitar espacios dobles
    s = " ".join(s.split())
    return s
correccion_bsas = {
    'Partidos del AGBA': 'buenos aires',
    'Provincia de Buenos Aires (sin AGBA)': 'buenos aires',
    'Ciudad Autonoma de Buenos Aires': 'caba'
}
df_bsas['provincia_nombre'] = df_bsas['provincia_nombre'].replace(correccion_bsas)

df_bsas['provincia_nombre'] = df_bsas['provincia_nombre'].apply(normalizar_textos)
df_bsas['departamento_nombre'] = df_bsas['departamento_nombre'].apply(normalizar_textos)
gdf['NOMPROV'] = gdf['NOMPROV'].apply(normalizar_textos)
gdf['NOMDEPTO'] = gdf['NOMDEPTO'].apply(normalizar_textos)
df_bsas = df_bsas[df_bsas['provincia_nombre'].isin(['buenos aires', 'caba'])]


# Extraer los Ãºltimos 8 dÃ­gitos de LINK
gdf['LINK_8'] = gdf['LINK'].astype(str).str[-8:]

# Pasar radio_codigo a string
df_bsas['radio_codigo'] = df_bsas['radio_codigo'].astype(str)

# Ver si coinciden
codigos_comunes = set(df_bsas['radio_codigo']).intersection(set(gdf['LINK_8']))

print("CÃ³digos comunes:", len(codigos_comunes))
print("Ejemplo de coincidencias:", list(codigos_comunes))

```
